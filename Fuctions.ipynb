```python
#map function
nums = [1,2,3,4,5]
def sq(n):
  return n*n
sq_nums = map(sq, nums)
print(list(sq_nums))
```

```python
#zip function
name = ["Henry", "Roy","Julie"]
ID = [4,1,2]
mapped = zip(name, ID)
print(list(mapped))
```
