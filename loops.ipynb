```python
# For loop
fruits = ["Apple", "Pineapple","Banana", "Cherry", "Orange"]
for x in fruits:
  print(x)
```

```python
#Break
for x in fruits:
  print(x)
  if x == "Banana":
    break
```

```python
#Continue
for x in fruits:
  if x == "Banana":
    continue
  print(x)
```

```python
#While loop
num = 1
while num < 10:
  print(num)
  num += 1
```

```python
num = 1
while num <11:
  print(num)
  if num == 7:
   break
  num+=1
```
