```python
#add()
fruits = {"apple", "banana", "cherry"}
fruits.add("orange")
print(fruits)
```

```python
#pop()
fruits = {"apple", "banana", "cherry", "orange"}
fruits.pop()
print(fruits)
```

```python
#remove()
fruits = {"apple", "banana", "cherry", "orange"}
fruits.remove("banana")
print(fruits)
```

```python
#difference
x={"apple", "banana", "cherry"}
y={"google", "microsoft", "apple"}
print(y.difference(x))
```

```python
#intersection()
x={"apple", "banana", "cherry"}
y={"google", "microsoft", "apple"}
print(x.intersection(y))
```

```python
#union()
x={"apple", "banana", "cherry"}
y={"google", "microsoft", "apple"}
print(x.union(y))
```

```python
#update()
x={"apple", "banana", "cherry"}
y={"google", "microsoft", "apple"}
y.update(x)
print(y)
```
