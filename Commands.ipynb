```python
import pandas as pd

# Define the student data
student_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Age': [20, 21, 19, 22, 20, 23, 21],
    'University': ['MIT', 'Stanford', 'Harvard', 'Yale', 'Princeton', 'Columbia', 'Duke']
}
df_students = pd.DataFrame(student_data)
output_path = '/content/data.csv'
df_students.to_csv(output_path, index=False)

print(f"Successfully created '{output_path}' with student records.")
display(df_students.head())
```

```python
student_data['City'] = ['Iris','Tokoyo','London','Spain','Paris','Berlin','Rome']
df_student = pd.DataFrame(student_data)
display(df_student)
```
