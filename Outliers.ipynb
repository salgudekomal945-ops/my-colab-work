```python
list_ = [1,2,3,4,5,6,9,10,12,55,60]
```

```python
import numpy as np
```

```python
Q1 = np.percentile(list_,25)
Q1
```

```python
Q3 = np.percentile(list_,75)
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
low = Q1 - 1.5 * IQR   #Lower quartile
low
```

```python
upp = Q3 + 1.5 * IQR   #Upper quartile
upp
```

```python
Outliers = []
```

```python
for i in list_:
    if ((i>upp) or (i<low)):
        Outliers.append(i)
```

```python
Outliers
```

```python
for i in list_:
  if i not in Outliers:
    print(i)
```

```python
list_ = [i for i in list_ if i not in Outliers]
```
