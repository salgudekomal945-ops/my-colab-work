```python
import pandas as pd
names = pd.DataFrame({'Name':['Tom','Jack','Steve','Ricky','Sam','Will','Root']})
ages = pd.DataFrame({'Age':[28,34,29,42,17,17,45]})
```

```python
names
```

```python
ages
```

```python
names2 = pd.DataFrame({'Name':['Joe','Pope','Buttler','Ben','James','Cook']})
```

```python
data = pd.concat([names,ages], axis=1) #Concatenation
display(data)
```

```python
pd.concat([names,names2]).reset_index()   #Concatenation
```

```python
display(data['Name'])
data.loc[1,'Age']
```

```python
#Series in pandas
series = pd.Series([1,2,7,8,5])
print(series)
```

```python
#Series with index
series = pd.Series([1,2,7,8,5], index=['a','b','c','d','e'])
print(series)
```

```python
#Accessing Series element
print(series["a"])
print(series["c"])
```

```python
#DataFrame method on list
data = pd.DataFrame([1,2,3,4,5])
print(data)
```

```python
#DataFrmae using Dictionary
data = pd.DataFrame({'Name':['Tom','Jack','Steve','Ricky','Sam','Will','Root'],
                  'Age':[28,34,29,42,17,17,43]  },index = ['C1','C2','C3','C4','C5','C6','C7'])
data.head(7)
```
