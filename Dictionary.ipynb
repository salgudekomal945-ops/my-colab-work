```python
car = {"brand": "Ford", "model":"Mustang", "year": "1964"}
#Get()method
car.get("model")

#items()method
car.items()

#key() method
car.keys()

#values() method
car.values()

#pop()method
car.pop("model")
print(car)

#popitem()method
car.popitem()
print(car)

#update()method
car.update({"color":"White"})
print(car)

#clear()method
car.clear()
print(car)
```
