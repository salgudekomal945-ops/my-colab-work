```python
#Standard methods
fruits = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(fruits)
```

```python
#Allows duplicates
fruits = ("apple", "banana", "cherry", "apple", "cherry")
print(fruits)
```

```python
#Count() method
nums = (2, 4, 2, 6, 2)
nums.count(2)
```

```python
#Index() method
nums = (2, 4, 2, 6, 2)
nums.count(4)
```
