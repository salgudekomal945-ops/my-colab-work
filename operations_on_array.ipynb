```python
import numpy as np
a = np.array([[1,2,3,],
             [4,5,6]])
#append method
np.append(a,[7,8,9])
np.append(a,[[7,8,9]],axis=0)
np.append(a,[[5],[8]],axis=1)
```

```python
a = np.array([[1,2],
              [3,4],
              [5,6]])
#insert method
np.insert(a,3,[7,8])
np.insert(a,1,[7,8],axis=0)
np.insert(a,2,[7,8,9],axis=1)
```

```python
a = np.array([[1,2,3,12],
              [9,5,6,11],
              [8,4,10,7]])
#delete method
np.delete(a,1)
np.delete(a,1,axis=0)
np.delete(a,1,axis= 1)
```

```python
np.ndim(a)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a,b)
print(np.ndim(a),np.ndim(b))
```

```python
#Hstack
result = np.hstack((a,b))
print(result, np.ndim(result))
```

```python
#Vstack
result = np.vstack((a,b))
print(result,np.ndim(result))
```

```python
#concatenate
arr1 = np.array([[1,2,3],
                 [4,5,6]])
arr2 = np.array([[7,8,9],
                 [10,11,12]])
new = np.concatenate((arr1,arr2),axis=0)  #adding rows
print(new)
```

```python
arr1 = np.array([[1,2,3],
                 [4,5,6]])
arr2 = np.array([[7,8,9],
                 [10,11,12]])
new = np.concatenate((arr1,arr2),axis=1)  #adding columns
print(new)
```
